```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np
np.random.seed(42)
#Make sure Cancer has higher gene values than Normal
data = pd.DataFrame({
    'Gene1':np.concatenate([np.random.normal(7,1,50),np.random.normal(3,1,50)]),
    'Gene2':np.concatenate([np.random.normal(6,1,50),np.random.normal(2,1,50)]),
    'Gene3':np.concatenate([np.random.normal(8,1,50),np.random.normal(4,1,50)]),
    'Label':['Cancer']*50 + ['Normal']*50})
#split data
x= data[['Gene1','Gene2','Gene3']]
y= data['Label']
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)

#Train ML model
model = SVC(kernel='linear')
model.fit(x_train,y_train)
#predict and check accuracy
y_pred = model.predict(x_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.1f}%")
print("\nYour ML model can distinguish Cancer vs Normal samples!")
```

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
#Load real cancer data
cancer_data = load_breast_cancer()
df=pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
df['Label'] = cancer_data.target
df['Label'] = df['Label'].map({0: 'Cancer', 1:'Normal'})

print(f"Real data loaded! Shape:{df.shape}")
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
#Split real data
x=df.drop('Label',axis=1)
y=df['Label']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
#Train model on real data
model=SVC(kernel='linear')
model.fit(x_train,y_train)
#Test accuracy
y_pred = model.predict(x_test)
print(f"Accuracy on Real cancer data:{accuracy_score(y_test, y_pred)*100:1f}%")
```
